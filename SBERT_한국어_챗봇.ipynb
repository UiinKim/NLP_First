```python
import numpy as np
import pandas as pd
from numpy import dot
from numpy.linalg import norm
import urllib.request
from  sentence_transformers import SentenceTransformer
```

```python
urllib.request.urlretrieve("https://raw.githubusercontent.com/songys/Chatbot_data/master/ChatbotData.csv", filename="ChatBotData.csv")
train_data=pd.read_csv('ChatBotData.csv')
```

```python
train_data.head()
```

```python
#pre-trained 된 BERT를 로드한다.
model=SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
#평균 풀링을 사용했으며 nli(자연어 추론)데이터를 학습후 sts(spring tool suite)데이터로 추가 파인튜닝한 모델
```

```python
#train_data의 Q에 대한 문자 임베딩 값을 구한 후 embedding열을 추가하여 저장
train_data['embedding']=train_data.apply(lambda row:model.encode(row.Q), axis=1)
```

```python
#코사인 유사도 함수
def cos_sim(A,B):
  return dot(A,B)/(norm(A)*norm(B))
```

```python
def return_answer(question):
  embedding=model.encode(question)
  #train_data의 score에는 사전 훈련된 qeuestion의 embedding값과 현재 받은 질문의 embedding값의 코사인 유사도를 구하고 저장한다.
  train_data['score']=train_data.apply(lambda x:cos_sim(x['embedding'],embedding), axis=1)
  return train_data.loc[train_data['score'].idxmax()]['A'] #train_data의 score중에서 max값을 가진 인덱스의 'A'의 값을 반환한다.
```

```python
return_answer('결혼하고싶어')
```

```python
return_answer('시간 있어?')
```

```python
return_answer('논문 작성 해줘')
```
